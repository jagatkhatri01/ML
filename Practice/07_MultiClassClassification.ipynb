```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.concat(
    map(pd.read_csv, ['Data/iris_train.csv', 'Data/iris_test.csv']), ignore_index=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
dataset.drop(columns='Unnamed: 5', inplace=True)
```

```python
dataset.head(2)
```

```python
sns.pairplot(data=dataset, hue='class')
plt.show()
```

```python
x = dataset.iloc[:, :-1]
y = dataset['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

# Normal

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
lr.score(X_test, y_test)*100
```

# OVR

```python
lr1 = LogisticRegression(multi_class='ovr')
lr1.fit(X_train, y_train)
```

```python
lr1.score(X_test, y_test)*100
```

# softmax

```python
lr2 = LogisticRegression(multi_class='multinomial')
lr2.fit(X_train, y_train)
```

```python
lr2.score(X_test, y_test)*100
```
