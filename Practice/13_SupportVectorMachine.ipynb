```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from mlxtend.plotting import plot_decision_regions
import seaborn as sns
```

```python
dataset = pd.read_csv('Data/placement-dataset.csv')
```

```python
dataset.isnull().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.shape
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x='cgpa', y='iq', hue='placement', data=dataset)
plt.show()
```

```python
X = dataset[['cgpa', 'iq']]  # Input/Features = Independent variables # must be multidimensional 2D
y = dataset['placement']  
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
sv = SVC(kernel='linear')
```

```python
sv.fit(X_train, y_train)
```

```python
y_pred = sv.predict(X_test)
```

```python
sv.score(X_test, y_test)*100
```

```python
sv.score(X_train, y_train)*100
```

```python
plot_decision_regions(X.to_numpy(), y.to_numpy(), clf=sv)
plt.show()
```
