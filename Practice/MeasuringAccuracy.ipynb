```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv('Data/iris_train.csv')
```

```python
data
```

```python
x = data.iloc[:, :-1]
x.head(2)
```

```python
y = data.iloc[:, -1]
y.head(2)
```

```python
rfc =  DecisionTreeClassifier()
```

```python
rfc.fit(x, y)
```

```python
predictions = rfc.predict([[6.3,	3.3,	6	, 2.5]])
```

```python
predictions
```

```python
dataTest = pd.read_csv('Data/iris_test.csv')
```

```python
x_test = dataTest.iloc[:, :-2]
x_test
```

```python
predictions = rfc.predict(x_test)
```

```python
predictions
```

```python
actual = dataTest.iloc[:, -2]
actual
```

```python
count = 0
for i in range(0, len(actual)):
    if actual[i]==predictions[i]:
        count += 1
print((count*100)/len(actual))
```
