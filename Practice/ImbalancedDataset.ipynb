```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
# pip install imblearn 
```

```python
dataset = pd.read_csv('Data/Social_Network_Ads.csv')
```

```python
dataset.head(20)
```

```python
x = dataset[['Age', 'EstimatedSalary']]
y = dataset['Purchased']
```

# THe data is imbalanced. I use Random under sampling to balanced the data. 

```python
y.value_counts()   # The data is imbalanced 
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
ru = RandomUnderSampler()
ru_x, ru_y = ru.fit_resample(x, y)   # Now use ru_x and ru_y as x and y in train test split and check score
```

```python
ru_x.head(2)
```

```python
ru_y.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(ru_x, ru_y, random_state=42, test_size=0.2)
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
lr.score(X_test, y_test)*100
```

```python
lr.predict([[45, 26000]])
```
