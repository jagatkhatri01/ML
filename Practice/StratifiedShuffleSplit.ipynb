```python
print('')
import numpy as np
def shuffle_and_split(data, test_ratio):
    np.random.seed(42) # Set the seed for reproducibility 
    shuffle_indices = np.random.permutation(len(data))  # THis returns shuffle indices
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffle_indices[:test_set_size]
    train_indices = shuffle_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]


# How the slice works: shuffle_indices[:test_set_size]

# shuffle_indices is an array: [5, 2, 8, 0, 3, ...]

# test_set_size is an integer: 4128

# In Python/NumPy, slicing an array with an integer works like this:

# array[:n]   # returns first n elements
```

```python
train, test = shuffle_and_split(df, 0.2)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
split = StratifiedShuffleSplit(n_splits=1, random_state=42, test_size=0.2)
for train_index, test_index in split.split(df, df['income_cat']):
    strata_train_set = df.loc[train_index]
    strata_test_set = df.loc[test_index]
```
