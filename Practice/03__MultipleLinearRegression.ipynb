```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# import sklearn.preprocessing import 
from sklearn.linear_model import LinearRegression
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = fetch_california_housing()
```

```python
data
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
df
```

```python
X = df
y = data.target
```

```python
print(type(X))
print(type(y))
```

```python
# X["MedianHouseValue"] = y
X.head()
```

```python
plt.figure(figsize=(10, 8))
import seaborn as sns 
sns.heatmap(X.corr(), annot=True)
```

```python
# x = data.data
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
y_test
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
y_pred
```

```python

X_test = X_test.values.reshape(-1, 1)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Squared Error", mse)
print(f"R2 score", r2)
print("\nFeature Weights (Importance):")
for name, w in zip(data.feature_names, reg.coef_):
    print(f"    {name}: {w:.4f}")
print(f"Intercept: {reg.intercept_:.2f}")
```

```python
plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred, alpha=0.6)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', lw=2)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual vs Predicted (Multiple Linear Reg)')
plt.show()
```

```python
plt.scatter(X_test, y_test, label='Actual')
plt.scatter(X_test, y_pred, label='Prediction')
plt.xlabel('Actual Price')
plt.ylabel('Predicted price')
plt.title('Linear Regression: multiple linear Reg')
plt.legend()
plt.show()
```
