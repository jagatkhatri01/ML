```python
import pandas as pd
import numpy as np
print('start')
```

```python
rng = np.random.default_rng(seed=42)
study_hours = rng.integers(low=1, high=11, size=50)
sleep_hours = rng.integers(low=4, high=10, size=50)
exam_score = (study_hours*5) + (sleep_hours*3) + np.random.normal(0, 5, 50)

df = pd.DataFrame({'StudyHours':study_hours,
                   'SleepHours':sleep_hours,
                   'ExamScore':exam_score})
print("=== Student Performance Dataset (Sample) ===")
print(df.head())
# print(df.head().to_string(index=True))
print()

avg_hour = study_hours.mean()
avg_sleep = sleep_hours.mean()
avg_score = exam_score.mean()

print("Dataset Summary: ")
print("Average study hours: ", avg_hour)
print("Average sleep hours: ", avg_sleep)
print(f"Average exam score: {avg_score:.1f}")

print("=== Prediction for New Student ===")
new_study = int(input("Enter study hours (1-10): "))
new_sleep = int(input("Enter sleep hours (4-9): "))

predicted = (avg_study * 5) + (avg_sleep * 3)      # <-- the "weights" rule
print(f"New Student: {new_study} study hours, {new_sleep} sleep hours")
print(f"Predicted Score (using averages): {predicted:.1f}")
```
