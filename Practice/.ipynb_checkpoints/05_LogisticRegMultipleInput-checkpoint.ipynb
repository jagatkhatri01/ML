```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv('Data/placement-dataset.csv')
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.head(2)
```

# The data is linearly separable. So, Logistic Regression is best for the data.

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x='cgpa', y='iq', hue='placement', data=dataset)
plt.show()
```

```python
X = dataset[['cgpa', 'iq']]  # Input/Features = Independent variables # must be multidimensional 2D
y = dataset['placement']  # Target/Labels. Output.
print(type(X))
print(X.shape)
print(type(y))
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

# Training the model

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
lr.score(X_test, y_test)*100
```

```python
lr.predict([[6.8	,123.0	]])
```

```python
lr.predict([[5.9	, 106.0]])
```

```python
plot_decision_regions(X.to_numpy(), y.to_numpy(), clf=lr)
plt.show()
```
