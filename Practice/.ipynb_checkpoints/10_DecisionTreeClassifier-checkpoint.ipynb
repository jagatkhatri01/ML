```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from mlxtend.plotting import plot_decision_regions
```

```python
data = pd.read_csv('Data/Social_Network_Ads.csv')
```

```python
data.head(3)
```

```python
sns.scatterplot(x='Age', y='EstimatedSalary', hue='Purchased', data=data)
plt.show()
```

```python
data.isnull().sum()
```

```python
data.columns
```

```python
x = data[['Age', 'EstimatedSalary']]
y = data['Purchased']
```

```python
import matplotlib.pyplot as plt
sns.countplot(x='Purchased', data=data)
plt.title('Distribution of Purchase (0 vs 1)')
plt.xlabel('Purchase')
plt.ylabel('Count')
plt.show()
```

# Scaling

```python
sc = StandardScaler()
```

```python
sc.fit(x)
```

```python
x_processed = pd.DataFrame(sc.transform(x), columns=x.columns)
```

```python
x_processed.head(2)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_processed, y, test_size=0.2, random_state=42)
```

```python

sns.countplot(x=y_train)
plt.title('Distribution of Purchase (0 vs 1)')
plt.xlabel('Purchase')
plt.ylabel('Count')
plt.show()

sns.countplot(x=y_test)
plt.title('Distribution of Purchase (0 vs 1)')
plt.xlabel('Purchase')
plt.ylabel('Count')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(X_train, y_train)
```

```python
dt.score(X_test, y_test)
```

```python
plt.figure(figsize=(10,6))
plot_decision_regions(x_processed.to_numpy(), y.to_numpy(), clf=dt)
plt.show()
```

```python
dt1 = DecisionTreeClassifier(criterion='entropy', max_depth=3) # After post-pruning
```

```python
dt1.fit(X_train, y_train)
```

```python
dt1.score(X_test, y_test)*100
```

```python
plt.figure(figsize=(10,6))
plot_decision_regions(x_processed.to_numpy(), y.to_numpy(), clf=dt)
plt.show()
```

# Pre-Pruning

```python
pre_dt = DecisionTreeClassifier(max_depth=5)
```

```python
pre_dt.fit(X_train, y_train)
```

```python
pre_dt.score(X_test, y_test)*100
```

```python
pre_dt.score(X_train, y_train)*100
```

```python
plt.figure(figsize=(10, 6))
plot_tree(pre_dt)
plt.show()
```

# Post-pruning

```python
for i in range(1, 20):
    post_dt = DecisionTreeClassifier(max_depth=i)
    post_dt.fit(X_train, y_train)
    print((post_dt.score(X_train, y_train)*100), (post_dt.score(X_test, y_test)*100), i)
```
