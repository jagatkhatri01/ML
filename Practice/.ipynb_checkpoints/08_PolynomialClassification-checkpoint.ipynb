```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import seaborn as sns
from sklearn.preprocessing import PolynomialFeatures
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv('Data/polynomial_classification.csv')
```

```python
dataset.head()
```

```python
dataset.columns
```

```python
sns.scatterplot(x='data1', y='data2', hue='output', data=dataset)
plt.show()
```

```python
X = dataset.iloc[:, :-1]
y = dataset['output']
```

```python
pf = PolynomialFeatures(degree=6)
pf.fit(X)
x_poly = pd.DataFrame(pf.transform(X))
print(x_poly.shape)
```

```python
# X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2) # Without polynomial
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_poly, y, random_state=42, test_size=0.2)
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
lr.score(X_test, y_test)*100
```
