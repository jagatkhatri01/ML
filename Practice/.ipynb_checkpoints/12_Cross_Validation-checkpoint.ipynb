```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
dataset = pd.read_csv('Data/placement-dataset.csv')
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
x = dataset[['cgpa']]
y = dataset['placement']
```

```python
from sklearn.model_selection import cross_val_score
```

```python
p = cross_val_score(RandomForestClassifier(), x, y, cv=5)
p.sort()
p*100
```
