```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Data/emissions.csv')
```

```python
x = df[['Fuel Consumption Comb (mpg)']]
y = df['Engine Size(L)']
print(x.shape, x.ndim)
print(y.shape, y.ndim)
```

```python
plt.scatter(x, y)
```

```python
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)
x_poly
```

```python
model = LinearRegression()
model.fit(x_poly, y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_poly, y, test_size=0.25, random_state=42)
```

```python
X_test, y_test
```

```python
model.score(X_test, y_test)
```

```python
train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)
print(f"Training R²: {train_score:.4f}")
print(f"Testing R²: {test_score:.4f}")
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# y = m1 x1 + m2 x1^2 + c
# y = -0.44142793 x1 + 0.00482885 x1^2 + 11.391778660547708
```

```python
x_range = np.linspace(x.min(), x.max(), 100).reshape(-1, 1)
# NO! The linspace data is ONLY for plotting - NOT for predictions!
x_range_poly = poly.transform(x_range)  # Transform using the same polynomial features
y_range_pred = model.predict(x_range_poly)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_train[:, 1], y_train, color='blue', alpha=0.5, label='Training Data')
# Plot test data  
plt.scatter(X_test[:, 1], y_test, color='green', alpha=0.5, label='Test Data')
# Plot the smooth regression curve
plt.plot(x_range, y_range_pred, color='red', linewidth=3, label='Polynomial Fit')

plt.xlabel('Fuel Consumption Comb (mpg)')
plt.ylabel('Engine Size(L)')
plt.title('Polynomial Regression (Degree=2) - Train/Test Split')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# 7. Print model equation
coefficients = model.coef_
intercept = model.intercept_
print(f"\nModel Equation: y = {coefficients[2]:.6f}x² + {coefficients[1]:.6f}x + {intercept:.6f}")
```

```python
# Get predictions for both train and test sets
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

plt.figure(figsize=(10, 6))

# Plot all original data
plt.scatter(x, y, color='gray', alpha=0.3, label='All Data')
# Plot training predictions
plt.scatter(X_train[:, 1], y_train_pred, color='blue', alpha=0.7, label='Train Predictions')
# Plot test predictions  
plt.scatter(X_test[:, 1], y_test_pred, color='green', alpha=0.7, label='Test Predictions')
# Plot smooth curve
plt.plot(x_range, y_range_pred, color='red', linewidth=2, label='Regression Curve')

plt.xlabel('Fuel Consumption Comb (mpg)')
plt.ylabel('Engine Size(L)')
plt.title('Polynomial Regression - Predictions vs Actual')
plt.legend()
plt.show()
```
