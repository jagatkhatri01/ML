```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Dataset/iris.csv')
```

```python
dataset.shape
```

```python
dataset.head(5)
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
import scipy.cluster.hierarchy as sc
```

```python
sc.dendrogram(sc.linkage())
```

```python
plt.figure(figsize=(20, 20))
sc.dendrogram(sc.linkage(dataset, method='single', metric='euclidean'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
agglo_clust = AgglomerativeClustering(n_clusters=2, linkage='single')
```

```python
dataset['predict'] = agglo_clust.fit_predict(dataset)
```

```python
sns.pairplot(data=dataset, hue = 'predict')
plt.show()
```
